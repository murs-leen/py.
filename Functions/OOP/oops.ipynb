Inheritance

```python
class Car:

    
    def __init__(self,type):
        self.type = type

    def start(self):
        print("Car is starting....")


    def stop(self):
        print("Car is stopping...")


class Toyota(Car):

    
    def __init__(self,brand,type):
        super().__init__(type) # calls constructor of parent
        self.brand = brand


    def display(self):
        print("Brand: ",self.brand)
        print("Type: ",self.type)
        self.start()
        self.stop()


car = Toyota("Toyota GLI","Four Wheler")
car.display()
    
```

```python
class Animal:

    
    def __init__(self):
        pass

        
    @staticmethod
    def sound():
        print("Some sound")


class Dog():

    
    @staticmethod
    def sound():
        Animal.sound()
        print("Bark")


a = Animal()
d = Dog()
a.sound()
d.sound()
```
